```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://boston.craigslist.org/search/sof"
```

```python
response = requests.get(url)
```

```python
response
```

```python
data = response.text
```

```python
soup= BeautifulSoup(data,'html.parser')
```

```python
tags= soup.find_all('a')
```

```python
for tag in tags:
    print(tag.get('href'))
```

```python
titles = soup.find_all("a",{"class":"result-title"})
```

```python
for title in titles:
    print(title.text)
```

```python
locations = soup.find_all("span",{"class":"result-hood"})
```

```python
for location in  locations:
    print(location.text)
```

```python
paragraphs = soup.find_all('p',{"class", "result-info"})
```

```python
for paragraph in paragraphs:
    title = paragraph.find('a',{"class":"result-title"}).text
    location_tag = paragraph.find("span",{"class":"result-hood"})
    location = location_tag.text[2: -1] if location_tag else "N/A"
    date = paragraph.find('time',{'class': 'result-date'}).text
    link = paragraph.find('a',{"class":"result-title"}).get("href")
    paragraph_response = requests.get(link)
    paragraph_data = paragraph_response.text
    paragraph_soup= BeautifulSoup(paragraph_data, 'html.parser')
    paragraph_description =paragraph_soup.find('section',{'id':'postingbody'}).text
    paragraph_attributes_tag = paragraph_soup.find('p',{'class':'attrgroup'})
    paragraph_attributes = paragraph_attributes_tag.text if paragraph_attributes_tag else 'N\A'
    print("Job Title:", title, '\nLocation', location, '\nDate:', date, '\nLink:', link ,"\n", paragraph_attributes, '\nJob Description', paragraph_description,  '\n----' )
    
    
```
